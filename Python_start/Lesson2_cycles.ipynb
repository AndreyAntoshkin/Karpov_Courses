## У нас есть словарь users, в котором хранится информация о группе пользователей: ключ — логин пользователя, значение — сколько у пользователя покупок.

### Мы хотим отобрать только тех пользователей, у которых число покупок выше среднего.

- Создайте множество above_average_users, в которое сохраните логины таких пользователей.



```python
# Пример 1
users = {'user1': 3, 'user2': 7, 'user3': 5}

avg = sum(users.values()) / len(users.values())

above_average_users = set()

for user, score in users.items():  # итерируемся по парам ключ-значение
    if score > avg:
        above_average_users.add(user)  # добавляем ключ (имя пользователя)


print(avg)
print(above_average_users)

# Ожидаемый результат
#above_average_users = {'user2'}
```

- Создайте переменную error_context, в которую сохраните словарь с контекстом вокруг ошибки, если он есть:
- Если есть событие до ошибки, добавьте ключ event_before со значением события.
- Если есть событие после ошибки, добавьте ключ event_after со значением события.
