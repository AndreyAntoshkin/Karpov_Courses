## У нас есть список prices, содержащий цены товаров. Создайте новый список updated_prices, в котором:

- Если цена превышает 450 рублей, она остаётся без изменений.
- Если цена меньше или равна 450 рублей, она увеличивается на 10% и округляется до одного знака после запятой.

```python
# Пример с дробной частью
prices = [99.9, 450, 451]

updated_prices = []

for price in prices:
    if price > 450:
        updated_prices.append(price)
    elif price <= 450:
        updated_prices.append(round(price * 1.1, 1))
 
print(updated_prices)
# Ожидаемый результат
#updated_prices = [109.9, 495.0, 451]
```

# При анализе отзывов пользователей часто важно понимать, сколько раз продукт получил наивысшую оценку. Это помогает оценить лояльность клиентов и качество сервиса.

## У вас есть список ratings, в котором хранятся оценки пользователей (целые числа от 1 до 5).

### Создайте две переменные:

- количество раз, когда в списке встречается максимальная оценка 5.
- count_above_three — количество оценок строго больше 3 (т.е. 4 или 5).

```python
# Пример 1

ratings = [5, 4, 5, 3, 5, 2]

count_five = 0
count_above_three = 0

for marks in ratings:
    if marks == 5:
        count_five += 1
    if marks > 3:
        count_above_three += 1
print(count_five)
print(count_above_three)

# Ожидаемый результат
#count_five = 3
#count_above_three = 4
```

# В аналитике часто требуется фильтровать данные из словаря по условию, например, выбирать пользователей с активностью выше среднего.

## У нас есть словарь users, в котором хранится информация о группе пользователей: ключ — логин пользователя, значение — сколько у пользователя покупок.

### Мы хотим отобрать только тех пользователей, у которых число покупок выше среднего.

- Создайте множество above_average_users, в которое сохраните логины таких пользователей.



```python
# Пример 1
users = {'user1': 3, 'user2': 7, 'user3': 5}

avg = sum(users.values()) / len(users.values())

above_average_users = set()

for user, score in users.items():  # итерируемся по парам ключ-значение
    if score > avg:
        above_average_users.add(user)  # добавляем ключ (имя пользователя)


print(avg)
print(above_average_users)

# Ожидаемый результат
#above_average_users = {'user2'}
```

# У нас есть список purchases, в котором сохранены все покупки пользователя за последнюю неделю.

## Создайте словарь products_count, в котором ключами будут все уникальные товары, а значениями — сколько раз они встречались в списке.



```python
# Пример 1
purchases = ['яблоко', 'банан', 'яблоко', 'груша']

product_count = dict()

for tovar in purchases:
    if tovar in product_count:
        product_count[tovar] +=1
    else:
        product_count[tovar] =1
print(product_count)

# Ожидаемый результат
#products_count = {'яблоко': 2, 'банан': 1, 'груша': 1}
```

При работе с веб-приложениями аналитики и разработчики часто сталкиваются с задачей валидации электронной почты и других данных пользователей.

В переменной email хранится строка с электронным адресом пользователя.

Проверьте, что e-mail удовлетворяет следующим условиям:

Символ '@' встречается ровно один раз.
До '@' есть хотя бы один символ.
После '@' указано название почтового ящика, затем точка и домен.
В строке нет пробелов.
E-mail не заканчивается точкой.
Создайте переменную is_valid, в которую сохраните True, если строка проходит все проверки, иначе False.

Примечание: конечно эти проверки не гарантируют, что перед нами стопроцентно работающий email, в этой задаче мы реализовали логику только самых базовых проверок. 

```python
email = 'user@example.com'

# Инициализация переменных
is_valid = True

# Проверка 1: Ровно один символ '@'
if email.count('@') != 1:
    is_valid = False

# Проверка 2: Есть хотя бы один символ до '@'
if is_valid and email.split('@')[0] == '':
    is_valid = False

# Проверка 3: После '@' есть точка и домен
if is_valid:
    after_at = email.split('@')[1]
    if '.' not in after_at or after_at.index('.') == 0 or after_at.index('.') == len(after_at) - 1:
        is_valid = False

# Проверка 4: Нет пробелов
if ' ' in email:
    is_valid = False

# Проверка 5: Не заканчивается точкой
if email.endswith('.'):
    is_valid = False

print(is_valid)
```
