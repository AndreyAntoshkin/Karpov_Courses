## В аналитике часто приходится хранить информацию о сущностях в виде словарей: пользователях, заказах, курьерах. Это удобно и позволяет быстро обращаться к нужным данным по ключу.

### Начнем с простого: создайте словарь с информацией о курьере нашего приложения.

#### Ваша задача: Создайте переменную courier, которая содержит словарь со следующими данными:

- имя: ключ name, значение "Иван"
- количество доставленных заказов: ключ delivered_orders, значение 125
- активен ли курьер в системе: ключ is_active, значение True

```python
courier = {
    'name': "Иван",
    'delivered_orders': 125,
    'is_active': True  
}

courier = dict(name = "Иван", delivered_orders = 125, is_active = True)

print(courier)
```

В аналитике часто требуется дополнять данные новыми полями. Например, при подготовке отчёта или формировании профиля пользователя.

В переменной user_info уже сохранён словарь с информацией о пользователе.

### Ваша задача:Добавьте в этот словарь два новых ключа:

- city со значением "Москва"
- is_premium со значением True
- Важно: не создавайте новый словарь, а обновляйте существующий.

### В словаре product_info сохранена информация о товаре. В словаре product_info_updated сохранена обновлённая информация про товар. Обновите данные в product_info данными из словаря product_info_updated таким образом, чтобы:

- в уже существующих ключах значения обновились,
- а новые ключи появились в словаре.
- В конце ваш код должен сохранить итоговый результат обратно в переменную product_info.

```python
 #Пример 1
product_info = {"id": 1, "name": "Наушники", "price": 2000}
product_info_updated = {"price": 1800, "in_stock": True}

# Ожидаемый результат
product_info = {"id": 1, "name": "Наушники", "price": 1800, "in_stock": True}



product_info.update(product_info_updated)
print(product_info)
```
