Напишите функцию calc_diff(numbers), которая принимает список чисел.

Функция должна возвращать разницу между максимальным и минимальным значением списка.

Если список пустой, функция возвращает None.

```python
def calc_diff(numbers):
    if not numbers:
        return None
    else:
        diff = max(numbers) - min(numbers)
        return diff

    
    
    

    
# Пример 1
print(calc_diff([]))
```

## Напишите функцию summarize_purchases(user_purchases), которая принимает словарь, где ключ — id пользователя (целое число), а значение — список id купленных товаров.

### Функция должна возвращать словарь с двумя ключами:

- "users": список уникальных id пользователей (ключи входного словаря), отсортированный по возрастанию.
- "purchases": список уникальных id всех покупок, отсортированный по возрастанию.

```python
def summarize_purchases(user_purchases):
    all_items = []
    for purchases in user_purchases.values():
        all_items.extend(purchases)
    return dict(users =  list(user_purchases.keys()), purchases = sorted(set(all_items)))

user_purchases = {
    1: [10, 20],
    2: [20, 30],
    3: [40]
}

summarize_purchases(user_purchases) 
```

Напишите функцию repeat_string(text, times), которая принимает строку text и целое число times.

Функция должна возвращать новую строку, в которой каждый символ исходной строки повторяется times раз подряд.

```python
def repeat_string(text, times):
    if times == 1:
        return text
    else:
        return ''.join(sorted(text * times))
    
#         result = ""
#     for char in text:
#         result += char * times
#     return result


# Пример 1
repeat_string("test", 1)
```

Напишите функцию analyze_activity(actions, threshold, include_low_activity), которая принимает три аргумента:

actions: список чисел — количество действий пользователя за каждый день.
threshold: число — пороговое значение.
include_low_activity: булево значение — учитывать ли дни с активностью ниже или равной порогу.
- Функция должна возвращать список индексов дней (нумерация с 0), которые удовлетворяют условию:

Если include_low_activity == True, в список попадают дни, когда количество действий меньше или равно threshold.
Если include_low_activity == False, в список попадают дни, когда количество действий больше threshold.

```python
def analyze_activity(actions, threshold, include_low_activity):
    result = []
    if include_low_activity:
        for idx, count in enumerate(actions):
            if count <= threshold:
                result.append(idx)
        return result
    else:
        for idx, count in enumerate(actions):
            if count > threshold:
                result.append(idx)
        return result



# Пример 1
#analyze_activity([5, 12, 8, 3], 5, True) #-> [0, 3]

# Пример 2
analyze_activity([5, 12, 8, 3], 5, False) #-> [1, 2]
```

Напишите функцию filter_users(users, min_purchases, active_only), которая принимает:

- users: словарь, где ключ — id пользователя (целое число), а значение — кортеж (количество_покупок, активен_ли_пользователь).
- min_purchases: минимальное количество покупок.
- active_only: булево значение, нужно ли выбирать только активных пользователей.
- Функция возвращает список id пользователей, которые:

- Если active_only=False, то функция возвращает список id пользователей, которые совершили не меньше min_purchases покупок.
- Если active_only=True, то функция возвращает список id пользователей, которые совершили не меньшеmin_purchases покупок И при  - этом были актиыными.  
- Список должен быть отсортирован по возрастанию id.

```python
def filter_users(users, min_purchases, active_only):
    result = []
    for user_id, user_data in users.items():
        purchases = user_data[0]
        is_active = user_data[1]
        if active_only:
            if purchases >= min_purchases and is_active is True:
                result.append(user_id)
        else:
            if purchases >= min_purchases:
                result.append(user_id)
    return(result)

users = {1: (5, True), 
         2: (3, False), 
         3: (7, True)}

# Пример 1
filter_users(users, min_purchases=4, active_only=True) #-> [1,3]

# так как стоит условие active_only=True будем отбирать только активных пользователей: 1 и 3
# у обоих активных пользователей число покупок больше min_purchases=4 
# поэтому функция возвращает список из двух id [1,3]


# Пример 2
#filter_users(users, min_purchases=1, active_only=False) -> [1, 2, 3]

# теперь active_only=False, значит можем не обращать внимание на то, активен ли пользователь
# у всех пользователей число покупок больше чем min_purchases=1
# поэтому функция возвращает список из трех id [1,2, 3]

# Пример 3
#filter_users(users, min_purchases=6, active_only=True) -> [3]
# вернули только одного активного пользователя, у которого прошло условие на минимальнео число покупок
```

Напишите функцию count_keyword_occurrences(reports, keyword), которая принимает:

- reports: словарь, где ключ — id пользователя, значение — список строк (отчётов).
- keyword: строка для поиска.
- Функция возвращает словарь, где ключ — id пользователя, а значение — сколько строк содержат keyword (регистр не учитывать).

```python
def count_keyword_occurrences(reports, keyword):
    slovar = {}
    sum_word = 0
    for user, stroka in reports.items():
        for message in stroka:
            if keyword.lower() in message.lower():
                sum_word +=1
        slovar[user] = sum_word
    return slovar

    summary = {}
    for user_id, texts in reports.items():
        count = 0
        for text in texts:
            if keyword.lower() in text.lower():
                count += 1
        summary[user_id] = count
    return summary



# Пример 1
reports = {1: ["Промокод save20 активирован", "Акция завершена"], 2: ["Промокод SAVE20 ещё действует"]}
count_keyword_occurrences(reports, "save20") #-> {1:1, 2:1}

# # Пример 2
# reports = {3: ["Нет данных", "Всё нормально"], 4: ["Промокод нет"]}
# count_keyword_occurrences(reports, "save20") -> {3:0, 4:0}

# # Пример 3
# reports = {5: ["Промокод save20 активирован"], 6: ["SAVE20 применён", "Другой отчёт"]}
# count_keyword_occurrences(reports, "save20") -> {5:1, 6:1}
```
