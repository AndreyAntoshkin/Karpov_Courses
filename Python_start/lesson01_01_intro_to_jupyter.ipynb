```python
# ‚ñ∂Ô∏è –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç –∑–∞–ø—É—Å–∫ —è—á–µ–µ–∫ –≤ Jupyter

# –ö–æ–≥–¥–∞ –≤—ã –ø–∏—à–µ—Ç–µ –∫–æ–¥ –≤ —è—á–µ–π–∫–µ ‚Äî –æ–Ω –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –ø–æ–∫–∞ –≤—ã –Ω–µ –Ω–∞–∂–∞–ª–∏ Ctrl+Enter –∏–ª–∏ Shift+Enter.
# –¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ –≤—ã –Ω–∞–ø–∏—Å–∞–ª–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é, –Ω–æ –Ω–µ –∑–∞–ø—É—Å—Ç–∏–ª–∏ —è—á–µ–π–∫—É ‚Äî –æ–Ω–∞ –Ω–µ –ø–æ–ø–∞–¥—ë—Ç –≤ –ø–∞–º—è—Ç—å, –∏ Python –Ω–µ –±—É–¥–µ—Ç –æ –Ω–µ–π –∑–Ω–∞—Ç—å.


# ‚ö†Ô∏è Jupyter –æ—Ç–∫–ª—é—á–∞–µ—Ç—Å—è, –µ—Å–ª–∏ –µ–≥–æ –¥–æ–ª–≥–æ –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å.
# –ü—Ä–∏ —Å–ª–µ–¥—É—é—â–µ–º –∑–∞—Ö–æ–¥–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏—Å—á–µ–∑–∞—é—Ç, –∏ –Ω—É–∂–Ω–æ —Å–Ω–æ–≤–∞ –∑–∞–ø—É—Å—Ç–∏—Ç—å —è—á–µ–π–∫–∏, —Å –∫–æ—Ç–æ—Ä—ã—Ö –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –≤–∞—à–∞ —Ä–∞–±–æ—Ç–∞.
```

```python
# ‚ûï –ß—Ç–æ–±—ã –¥–æ–±–∞–≤–∏—Ç—å –Ω–æ–≤—É—é —è—á–µ–π–∫—É:
# - –Ω–∞–∂–º–∏—Ç–µ –∫–Ω–æ–ø–∫—É "+" —Å–≤–µ—Ä—Ö—É
# - –∏–ª–∏ –Ω–∞–∂–º–∏—Ç–µ Esc ‚Üí B (–¥–æ–±–∞–≤–∏—Ç—å —Å–Ω–∏–∑—É) –∏–ª–∏ A (–¥–æ–±–∞–≤–∏—Ç—å —Å–≤–µ—Ä—Ö—É) –≤ —Ä–µ–∂–∏–º–µ –∫–æ–º–∞–Ω–¥

# –í –Ω–æ–≤–æ–π —è—á–µ–π–∫–µ –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –Ω–∞–ø–∏—Å–∞—Ç—å —á—Ç–æ-–Ω–∏–±—É–¥—å –≤—Ä–æ–¥–µ:
# print("–ê–Ω–∞–ª–∏—Ç–∏–∫ –¥–∞–Ω–Ω—ã—Ö!")
```

```python
# üíæ Jupyter –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –≤–∞—à—É —Ä–∞–±–æ—Ç—É,
# –Ω–æ –≤—ã –º–æ–∂–µ—Ç–µ —Ç–∞–∫–∂–µ –Ω–∞–∂–∞—Ç—å Ctrl+S (–∏–ª–∏ File ‚Üí Save Notebook)

# –≠—Ç–æ –æ—Å–æ–±–µ–Ω–Ω–æ –≤–∞–∂–Ω–æ, –µ—Å–ª–∏ –≤—ã —Ä–∞–±–æ—Ç–∞–µ—Ç–µ –≤ –±—Ä–∞—É–∑–µ—Ä–µ!
```

```python
# üü• –ö–Ω–æ–ø–∫–∞ "–°—Ç–æ–ø" (‚óº) ‚Äî –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∫–æ–¥–∞
# –ò–Ω–æ–≥–¥–∞ —è—á–µ–π–∫–∞ "–∑–∞–≤–∏—Å–∞–µ—Ç", –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ —Å–ª–∏—à–∫–æ–º –¥–æ–ª–≥–æ–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –∏–ª–∏ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ–º —Ü–∏–∫–ª–µ.
# –ï—Å–ª–∏ –≤—ã –≤–∏–¥–∏—Ç–µ, —á—Ç–æ —è—á–µ–π–∫–∞ –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç ‚Äî –Ω–∞–∂–º–∏—Ç–µ "–°—Ç–æ–ø", —á—Ç–æ–±—ã –ø—Ä–µ—Ä–≤–∞—Ç—å –µ—ë –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ.
```

```python
# üîÑ –ö–Ω–æ–ø–∫–∞ "Restart" ‚Äî –ø–µ—Ä–µ–∑–∞–ø—É—Å–∫–∞–µ—Ç Python –≤–Ω—É—Ç—Ä–∏ –Ω–æ—É—Ç–±—É–∫–∞
# –≠—Ç–æ –∫–∞–∫ "–ø–µ—Ä–µ–∑–∞–≥—Ä—É–∑–∫–∞ –ø–∞–º—è—Ç–∏": –≤—Å–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏—Å—á–µ–∑–∞—é—Ç, –≤—Å—ë –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –Ω—É–ª—è.

# –ö–æ–≥–¥–∞ —ç—Ç–æ –Ω—É–∂–Ω–æ?
# - –µ—Å–ª–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ "–∑–∞–≤–∏—Å–ª–∏" –≤ —Å—Ç—Ä–∞–Ω–Ω–æ–º —Å–æ—Å—Ç–æ—è–Ω–∏–∏
# - –∏–ª–∏ –ø—Ä–æ—Å—Ç–æ —Ö–æ—Ç–∏—Ç–µ –Ω–∞—á–∞—Ç—å —á–∏—Å—Ç–æ, –∫–∞–∫ –±—É–¥—Ç–æ —Ç–æ–ª—å–∫–æ –æ—Ç–∫—Ä—ã–ª–∏ –Ω–æ—É—Ç–±—É–∫

# –ü–æ—Å–ª–µ Restart –Ω—É–∂–Ω–æ –∑–∞–Ω–æ–≤–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —è—á–µ–π–∫–∏ —Å –∫–æ–¥–æ–º —Å–≤–µ—Ä—Ö—É!
```

```python
# –≠—Ç–∞ —è—á–µ–π–∫–∞ ‚Äî –∫–æ–¥–æ–≤–∞—è, –Ω–æ –±—ã–≤–∞—é—Ç –∏ —Ç–µ–∫—Å—Ç–æ–≤—ã–µ (Markdown) —è—á–µ–π–∫–∏.
# –í –Ω–∏—Ö –º–æ–∂–Ω–æ –ø–∏—Å–∞—Ç—å –∑–∞–≥–æ–ª–æ–≤–∫–∏, —Å–ø–∏—Å–∫–∏, –∂–∏—Ä–Ω—ã–π —Ç–µ–∫—Å—Ç –∏ —Ñ–æ—Ä–º—É–ª—ã.
# –ß—Ç–æ–±—ã –ø–µ—Ä–µ–∫–ª—é—á–∏—Ç—å —Ç–∏–ø —è—á–µ–π–∫–∏, –Ω–∞–∂–º–∏—Ç–µ "Code" –≤ –≤–µ—Ä—Ö–Ω–µ–º –º–µ–Ω—é –∏ –≤—ã–±–µ—Ä–∏—Ç–µ "Markdown".

# –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–æ–∑–¥–∞—Ç—å —Ç–µ–∫—Å—Ç–æ–≤—É—é —è—á–µ–π–∫—É –∏ –Ω–∞–ø–∏—Å–∞—Ç—å:
# ## –ú–æ–π –ø–µ—Ä–≤—ã–π –∑–∞–≥–æ–ª–æ–≤–æ–∫
# –≠—Ç–æ –º–æ–π –ø–µ—Ä–≤—ã–π —Ç–µ–∫—Å—Ç –≤ –Ω–æ—É—Ç–±—É–∫–µ.
```
